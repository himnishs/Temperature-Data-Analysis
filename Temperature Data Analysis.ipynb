```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
df = pd.read_csv("Temperatures/GlobalLandTemperaturesByMajorCity.csv") #Reads CSV File
data = df.groupby('City')['AverageTemperature'].mean().sort_values() #Sorts Values in Ascending Order
data.plot.bar() #Graphs it
plt.tick_params(axis='x', which='major', labelsize=3) #Makes it slighty readable
plt.tight_layout()
```

```python
harbin = df[df.City == 'Harbin'] #Gets all the rows with Harbin
Umm_Durman = df[df.City == 'Umm Durman'] #Gets all the rows with Umm Durman

date_truncate_harbin = harbin.dt.str.slice(0,4).drop_duplicates(inplace = False) # Gets Unique Dates
average_harbin = harbin.groupby(date_truncate_harbin)['AverageTemperature'].mean() #gets the mean of the date by using groupby

date_truncate_umm_durman = Umm_Durman.dt.str.slice(0,4).drop_duplicates(inplace = False) # Concates the data string into a year
average_umm_durman = Umm_Durman.groupby(date_truncate_umm_durman)['AverageTemperature'].mean() # gets the mean of the date by using groupby

plt.plot(date_truncate_umm_durman[145:155],average_umm_durman[145:155], label = "Umm Durman")
plt.plot(date_truncate_harbin[184:194], average_harbin[184:194], label = "Harbin")
plt.grid(True) #shows grid
plt.xlabel(" Year")
plt.ylabel(" Temperature(In Celsius)")
plt.title("The Hottest and Coldest Cities Compared from 2004-2013 (Umm Durman, Somalia), (Harbin, China)")
plt.legend() #Shows labels
plt.show()
```
